## Profit Return (%)

```python
def profit_return_1(df, days_after=5):
    df2 = df.copy()
    df2['Return % 5 days after']= " "
    for i in range(len(df2)- days_after):  
        if ( df2.iloc[i,4] == 1 ):
            df2.iloc[i,5] = ( ( df2.iloc[i+days_after, 3] - df2.iloc[i+1,0] )/df2.iloc[i+1,0] )*100
    
    return df2
```

```python
def profit_return_2(df, days_after=14):
    df2 = df.copy()
    df2['Return % 14 days after']= " "
    for i in range(len(df2) - days_after):  
        if ( df2.iloc[i,4] == 1 ):
            df2.iloc[i,6] = ( ( df2.iloc[i+days_after, 3] - df2.iloc[i+1,0] )/df2.iloc[i+1,0] )*100
    return df2
```

```python
def profit_return_3(df, days_after=28):
    df2 = df.copy()
    df2['Return % 28 days after']= " "
    for i in range(len(df2)-days_after):  
        if ( df2.iloc[i,4] == 1 ):
            df2.iloc[i,7] = ( ( df2.iloc[i+days_after, 3] - df2.iloc[i+1,0] )/df2.iloc[i+1,0] )*100
    return df2


#How to add ignore error
```

## Hammer Sample Collection

```python
def hammer_scanner(df, min_wick_ratio=5, min_candle_size=0.4):
    
    df2 = df.copy()
    
    #Adding columns
    df2['Hammer']=" "
    
    for i in range(len(df2)):
        cond1 = df2.iloc[i,[0,3]].min() - df2.iloc[i,2] > (df2.iloc[i,1] - df2.iloc[i,[0,3]].max())*min_wick_ratio
        cond2 = df2.iloc[i,[0,3]].min() > (df2.iloc[i,2]+((df2.iloc[i,1]-df2.iloc[i,2])/2))
        cond3 = df2.iloc[i,2] < df2.iloc[i-10:i,2].min()
        cond4 = abs(df2.iloc[i,0]-df2.iloc[i,3])<((df2.iloc[i,1]-df2.iloc[i,2])*min_candle_size)

        if ( ( cond1 == True )  and ( cond2 == True ) and ( cond3 == True ) and (cond4 == True) ):
                df2.iloc[i,-1] = 1
                
        else: 
                df2.iloc[i,-1] = 0
                
    
    return df2
```

```python
hammer_samples=pd.DataFrame()
for company in company_info:
    name=company[0]
    print("Fetching data for company", name)
    
    #Pass every copmpany dataframe through hammer scanner
    filtered_df=hammer_scanner(ndl_company_data_dict[name]['data'])
    filtered_df=profit_return_1(filtered_df)
    filtered_df=profit_return_2(filtered_df)
    filtered_df=profit_return_3(filtered_df)
    filtered_df=filtered_df[filtered_df['Hammer']==1]  
    
    hammer_samples=pd.concat([hammer_samples,filtered_df],axis=0)
```

## Shooting Star Sample Collection

```python
def shooting_star_scanner(df, min_wick_ratio=5, min_candle_size=0.4):
    
    df2 = df.copy()
    
    df2['Shooting Star']=" "
        
    for i in range(len(df2)):
        cond1 = ( df2.iloc[i,1] > df2.iloc[i-10:i,1].max() )
        x =  df2.iloc[i,1] - df2.iloc[i,[0,3]].max() 
        y = ( df2.iloc[i,[0,3]].min() - df2.iloc[i,2] ) * min_wick_ratio 
        cond2 = ( x > y ) 
        cond3 =  abs( df2.iloc[i,0] - df2.iloc[i,3] ) <  ( df2.iloc[i,1] - df2.iloc[i,2] ) * min_candle_size 

        if ( ( cond1 == True )  and ( cond2 == True ) and ( cond3 == True ) ):
                df2.iloc[i,-1] = 1
        else: 
                df2.iloc[i,-1] = 0

    return df2
```

```python
shooting_star_samples=pd.DataFrame()
for company in company_info:
    name=company[0]
    print("Fetching data for company", name)
    
    #Pass every copmpany dataframe through hammer scanner
    filtered_df_2=shooting_star_scanner(ndl_company_data_dict[name]['data'])
    filtered_df_2=profit_return_1(filtered_df_2)
    filtered_df_2=profit_return_2(filtered_df_2)
    filtered_df_2=profit_return_3(filtered_df_2)
    filtered_df_2=filtered_df_2[filtered_df_2['Shooting Star']==1]  
    display(filtered_df_2)
    shooting_star_samples=pd.concat([shooting_star_samples,filtered_df_2],axis=0)
```
